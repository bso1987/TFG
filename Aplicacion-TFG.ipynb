```python
import tkinter as tk

def move_to_in_progress():
    selected = todo_listbox.curselection()
    if selected:
        task = todo_listbox.get(selected)
        in_progress_listbox.insert(tk.END, task)
        todo_listbox.delete(selected)

def move_to_done():
    selected = in_progress_listbox.curselection()
    if selected:
        task = in_progress_listbox.get(selected)
        done_listbox.insert(tk.END, task)
        in_progress_listbox.delete(selected)

root = tk.Tk()
root.title("Trello Prototype")

# Frame for task input section
input_frame = tk.Frame(root)
input_frame.pack(side=tk.LEFT, padx=10)

task_label = tk.Label(input_frame, text="Task:")
task_label.pack()

task_entry = tk.Entry(input_frame, width=30)
task_entry.pack()

current_metric_label = tk.Label(input_frame, text="Current Metric:")
current_metric_label.pack()

current_metric_entry = tk.Entry(input_frame, width=30)
current_metric_entry.pack()

target_metric_label = tk.Label(input_frame, text="Target Metric:")
target_metric_label.pack()

target_metric_entry = tk.Entry(input_frame, width=30)
target_metric_entry.pack()

add_button = tk.Button(
    input_frame,
    text="Add Task",
    command=lambda: todo_listbox.insert(
        tk.END,
        f"Task: {task_entry.get()} - Current Metric: {current_metric_entry.get()} - Target Metric: {target_metric_entry.get()}"
    ),
)
add_button.pack(pady=5)

# Frame for To Do column
todo_frame = tk.Frame(root, width=600, height=400, bd=1, relief=tk.SOLID)
todo_frame.pack(side=tk.LEFT, padx=10, pady=10)

todo_label = tk.Label(todo_frame, text="To Do")
todo_label.pack()

todo_listbox = tk.Listbox(todo_frame, width=30, height=10)
todo_listbox.pack(pady=10, padx=5)

# Frame for In Progress column
in_progress_frame = tk.Frame(root, width=600, height=400, bd=1, relief=tk.SOLID)
in_progress_frame.pack(side=tk.LEFT, padx=10, pady=10)

in_progress_label = tk.Label(in_progress_frame, text="In Progress")
in_progress_label.pack()

in_progress_listbox = tk.Listbox(in_progress_frame, width=30, height=10)
in_progress_listbox.pack(pady=10, padx=5)

move_to_in_progress_button = tk.Button(in_progress_frame, text="Move to In Progress", command=move_to_in_progress)
move_to_in_progress_button.pack(pady=5)

# Frame for Done column
done_frame = tk.Frame(root, width=600, height=400, bd=1, relief=tk.SOLID)
done_frame.pack(side=tk.LEFT, padx=10, pady=10)

done_label = tk.Label(done_frame, text="Done")
done_label.pack()

done_listbox = tk.Listbox(done_frame, width=30, height=10)
done_listbox.pack(pady=10, padx=5)

move_to_done_button = tk.Button(done_frame, text="Move to Done", command=move_to_done)
move_to_done_button.pack(pady=5)

root.mainloop()


```
