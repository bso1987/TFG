```python
import tkinter as tk
from tkinter import messagebox

def move_task(source_listbox, destination_listbox):
    selected = source_listbox.curselection()
    if selected:
        task = source_listbox.get(selected)
        destination_listbox.insert(tk.END, task)
        source_listbox.delete(selected)

def add_metric():
    task = task_entry.get()
    current_metric = current_metric_entry.get()
    target_metric = target_metric_entry.get()
    
    if task and current_metric and target_metric:
        todo_listbox.insert(tk.END, f" {task} - Valor Actual: {current_metric} - Valor Objetivo: {target_metric}")
    else:
            # Mostrar mensaje de error si algún campo está vacío
        error_message = "Todos los campos son obligatorios. Por favor, complete todos los campos."
        messagebox.showerror("Error", error_message)

def update_metric():
    selected = todo_listbox.curselection()
    if selected:
        task = todo_listbox.get(selected)
        current_metric = current_metric_entry.get()
        target_metric = target_metric_entry.get()
        updated_task = f"{task.split('-')[0]} - Valor Actual: {current_metric} - Valor Objetivo: {target_metric}"
        todo_listbox.delete(selected)
        todo_listbox.insert(selected, updated_task)

app = tk.Tk()
app.title("Metrics Management App")

# Frame for task input section
input_frame = tk.Frame(app)
input_frame.pack(side=tk.LEFT, padx=10)

task_label = tk.Label(input_frame, text="Nombre de la métrica:")
task_label.pack()

task_entry = tk.Entry(input_frame, width=30)
task_entry.pack()

current_metric_label = tk.Label(input_frame, text="Valor actual:")
current_metric_label.pack()

current_metric_entry = tk.Entry(input_frame, width=30)
current_metric_entry.pack()

target_metric_label = tk.Label(input_frame, text="Valor objetivo:")
target_metric_label.pack()

target_metric_entry = tk.Entry(input_frame, width=30)
target_metric_entry.pack()

add_button = tk.Button(input_frame, text="Añadir Métrica", command=add_metric)
add_button.pack(pady=5)

update_button = tk.Button(input_frame, text="Actualizar Métrica", command=update_metric)
update_button.pack(pady=5)

# Frame for To Do column
todo_frame = tk.Frame(app, width=200, height=400, bd=1, relief=tk.SOLID)
todo_frame.pack(side=tk.LEFT, padx=10, pady=10)

todo_label = tk.Label(todo_frame, text="Métricas Objetivo")
todo_label.pack()

todo_listbox = tk.Listbox(todo_frame, width=50, height=10)
todo_listbox.pack(pady=10, padx=5)

move_to_todo_button = tk.Button(todo_frame, text="Iniciar seguimiento", command=lambda: move_task(in_progress_listbox, todo_listbox))
move_to_todo_button.pack(pady=5)

# Frame for In Progress column
in_progress_frame = tk.Frame(app, width=200, height=400, bd=1, relief=tk.SOLID)
in_progress_frame.pack(side=tk.LEFT, padx=10, pady=10)

in_progress_label = tk.Label(in_progress_frame, text="Métricas bajo seguimiento")
in_progress_label.pack()

in_progress_listbox = tk.Listbox(in_progress_frame, width=50, height=10)
in_progress_listbox.pack(pady=10, padx=5)

move_to_in_progress_button = tk.Button(in_progress_frame, text="Completar seguimiento", command=lambda: move_task(todo_listbox, in_progress_listbox))
move_to_in_progress_button.pack(pady=5)

# Frame for Done column
done_frame = tk.Frame(app, width=200, height=400, bd=1, relief=tk.SOLID)
done_frame.pack(side=tk.LEFT, padx=10, pady=10)

done_label = tk.Label(done_frame, text="Objetivo conseguidos")
done_label.pack()

done_listbox = tk.Listbox(done_frame, width=50, height=10)
done_listbox.pack(pady=10, padx=5)

move_to_done_button = tk.Button(done_frame, text="Retomar seguimiento", command=lambda: move_task(in_progress_listbox, done_listbox))
move_to_done_button.pack(pady=5)

app.mainloop()



```
